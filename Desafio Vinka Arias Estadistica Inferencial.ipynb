a. Realiza 5 pruebas con muestras de tamaño n = 25, en cada caso, con niveles de significancia diferentes pero inferiores a 0,1, para confirmar o rechazar la siguientes hipótesis considerando las alternativa 
 <left></h2>><left>



```python

# Muestras de tamaño 25
muestra = dataset_poblacion.sample(n=25, random_state=42)
muestra
```

```python

def prueba_hipotesis_media(datos, mu, colas):
    datos = datos.dropna()  # Eliminar valores NaN, por si acaso
    alpha=round(np.random.uniform(0.01, 0.1),2)
    confianza = 100 * (1 - alpha)
    
    t_stat, p_value = stats.ttest_1samp(datos, mu, alternative=colas)
    print(f"Con nivel de significancia de {alpha}.")
    if p_value < alpha:
        if colas == 'two-sided':
            print(f"Con nivel de confianza de {confianza}% se rechaza la hipótesis nula: la media es distinta de {mu}.")
        elif colas == 'greater':
            print(f"Con nivel de confianza de {confianza}% se rechaza la hipótesis nula: la media es mayor que {mu}.")
        elif colas == 'less':
            print(f"Con nivel de confianza de {confianza}% se rechaza la hipótesis nula: la media es menor que {mu}.")
    else:
        print(f"Con nivel de confianza de {confianza}% no se puede rechazar la hipótesis nula: la media es igual a {mu}.")
    
    print(f"Valor t: {t_stat}")
    print(f"Valor p: {p_value}\n")
```

```python

# Probar con diferentes valores de medias
#medias = [media - 50, media - 1, media, media + 50, media + 3000]

# Lista de tipos de prueba
tipos_de_prueba = ['two-sided', 'greater', 'less']
casos = {'earn':[[20000,'two-sided'],
                 [35000,'two-sided'],
                 [25000,'greater'],
                 [32000,'less']
                ],
         'height': [[55,'two-sided'],
                 [68,'two-sided'],
                 [63,'greater'],
                 [70,'less']
                ], 
          'age':[[44,'two-sided'],
                 [56,'two-sided'],
                 [51,'greater'],
                 [62,'less']
                ]}


## Recorrer el diccionario de casos y realizar las pruebas de hipótesis
for variable, pruebas in casos.items():
    for prueba in pruebas:
        valor_hipotetico, tipo_prueba = prueba
        print(f"Prueba para {variable} con H0: media = {valor_hipotetico}, tipo de prueba: {tipo_prueba}")
        prueba_hipotesis_media(muestra[variable], valor_hipotetico, tipo_prueba)
```
